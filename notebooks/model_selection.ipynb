```python
import numpy as np
import pandas as pd

from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression,Lasso, Ridge
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("feature_selection_data_outlier.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.shape
```

```python
x= df.drop(columns=['price'])
y=df['price']
```

```python
y_transformed = np.log1p(y)
```

```python
x.columns
```

### Ordinal Encoder

```python
enc_cols = ['facing','furnish_type','prop_type','city','ageing','location_area','floor_category','luxury_category']
```

```python
preprocessor=ColumnTransformer(
    [('num', StandardScaler(),['carpet_area','bedrooms','bathrooms','balcony','Store Room']),
     ('cat', OrdinalEncoder(handle_unknown="use_encoded_value",unknown_value=-1),enc_cols)],
    remainder='passthrough'
)
```

```python
pipeline = Pipeline(
    [('preprocessor',preprocessor),
     ('regressor',LinearRegression())]
)
```

```python
# K-fold cross-validation
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_val_score(pipeline, x, y_transformed, cv=kfold, scoring='r2')
```

```python
scores.mean(),scores.std()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y_transformed,test_size=0.2,random_state=42)
```

```python
pipeline.fit(x_train,y_train)
```

```python
y_pred = pipeline.predict(x_test)
mean_absolute_error(y_test,y_pred)
```

```python
y_pred=np.expm1(y_pred)
mean_absolute_error(np.expm1(y_test),y_pred)
```

```python
def scoring(model_name, model, preprocessor):

    output = []

    output.append(model_name)

    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])

    # K-fold cross-validation
    kfold = KFold(n_splits=10, shuffle=True, random_state=42)
    scores = cross_val_score(pipeline, x, y_transformed, cv=kfold, scoring='r2')

    output.append(scores.mean())

    X_train, X_test, y_train, y_test = train_test_split(x,y_transformed,test_size=0.2,random_state=42)

    pipeline.fit(X_train,y_train)

    y_pred = pipeline.predict(X_test)

    y_pred = np.expm1(y_pred)

    output.append(mean_absolute_error(np.expm1(y_test),y_pred))

    return output
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.neural_network import MLPRegressor
import xgboost
```

```python
model_dict = {
    'linear_reg':LinearRegression(),
    'svr':SVR(),
    'ridge':Ridge(),
    'LASSO':Lasso(),
    'decision tree': DecisionTreeRegressor(),
    'random forest':RandomForestRegressor(),
    'extra trees': ExtraTreesRegressor(),
    'gradient boosting': GradientBoostingRegressor(),
    'adaboost': AdaBoostRegressor(),
    'mlp': MLPRegressor(),
    'xgboost':xgboost.XGBRegressor()
}
```

```python
model_op = []
for model_name, model in model_dict.items():
    model_op.append(scoring(model_name,model,preprocessor))
```

```python
model_op
```

```python
model_df = pd.DataFrame(model_op,columns=['name','r2','mae'])
model_df.sort_values('mae')
```

### Hyperparameter Tuning

```python
enc_cols = ['facing','furnish_type','prop_type','city','ageing','location_area','floor_category','luxury_category']
```

```python
preprocessor=ColumnTransformer(
    [('num', StandardScaler(),['carpet_area','bedrooms','bathrooms','balcony','Store Room']),
     ('cat', OrdinalEncoder(handle_unknown="use_encoded_value",unknown_value=-1),enc_cols),
     ('cat1', OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False),['city','ageing']),
     ('target_enc',ce.TargetEncoder(),['location_area'])],
    remainder='passthrough'
)
```

```python
pipeline= Pipeline([
    ('preprocessor',preprocessor),
    ('regressor',RandomForestRegressor())
])
```

```python
kfold= KFold(n_splits=10, shuffle=True, random_state=42)
```

```python
import optuna
def objective(trial) :
  regressor__n_estimators = trial.suggest_int('regressor__n_estimators',50,750)
  regressor__criterion = trial.suggest_categorical('regressor__criterion',['squared_error', 'absolute_error', 'friedman_mse', 'poisson'])
  regressor__max_features = trial.suggest_categorical('regressor__max_features',['sqrt', 'log2', None])
  regressor__max_depth = trial.suggest_int('regressor__max_depth',1,50)
  regressor__min_samples_split = trial.suggest_int('regressor__min_samples_split',2,30)
  regressor__bootstrap = trial.suggest_categorical("regressor__bootstrap", [True, False])
  regressor__min_impurity_decrease = trial.suggest_float("regressor__min_impurity_decrease", 0.0, 1.0)

  # max_samples cannot be set if bootstrap is False
  if regressor__bootstrap:
      regressor__max_samples = trial.suggest_float("regressor__max_samples", 0.3, 1.0)
  else:
      regressor__max_samples = None  

  pipeline= Pipeline([
    ('preprocessor',preprocessor),
    ('regressor',RandomForestRegressor(n_estimators=regressor__n_estimators, max_depth=regressor__max_depth,
                                       min_samples_split=regressor__min_samples_split,bootstrap=regressor__bootstrap,
                                       min_impurity_decrease=regressor__min_impurity_decrease,
                                       criterion=regressor__criterion,max_features=regressor__max_features,
                                       max_samples = regressor__max_samples
                                       ))
  ])
  score= cross_val_score(pipeline,x,y_transformed,cv=kfold,scoring='r2')
  return score.mean()
```

```python
pruner = optuna.pruners.MedianPruner(n_warmup_steps=5)
study = optuna.create_study(direction='maximize',pruner=pruner)
sampler = optuna.samplers.TPESampler(seed=42)
study.optimize(objective, n_trials=120)
```

```python
# best_params = {k.replace("regressor__", ""): v for k, v in study.best_trial.params.items()}

best_params = {
    "n_estimators": 586,
    "criterion": "friedman_mse",
    "max_features": "log2",
    "max_depth": 44,
    "min_samples_split": 6,
    "bootstrap": False,
    "min_impurity_decrease": 0.00032896297305563165
}

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(**best_params))
])
```

```python
pipeline.fit(x,y_transformed)
```
